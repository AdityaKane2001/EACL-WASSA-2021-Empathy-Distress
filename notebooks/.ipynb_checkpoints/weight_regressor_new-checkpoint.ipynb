```python

nlp = spacy.load('en')

def get_lemmetized(df):
  # Convert to list 
  data = list(df['essay'])
  #print(data)
  #Remove Emails 
  data = [re.sub('\S*@\S*\s?', '', sent) for sent in data]  
  # Remove new line characters 
  data = [re.sub('\s+', ' ', sent) for sent in data]  
  # Remove distracting single quotes 
  data = [re.sub("\'", "", sent) for sent in data]  

  data_words = []
  essays_list_new = []
  for doc in data :
    essay_words_lem = []
    essay_new = ""
    doc = nlp(doc)
    for word in doc:
      if (word.lemma_.isalpha()) and  (word.lemma_ != '-PRON-'):
        essay_words_lem.append(word.lemma_.lower())
        essay_new = essay_new + " " + word.lemma_.lower()
    data_words.append(essay_words_lem)
    essays_list_new.append(essay_new)
  return data_words, essays_list_new


data_w, essay_new = get_lemmetized(df)
data_words, essay_new = get_lemmetized(df)
data_words_dev, essay_new_dev = get_lemmetized(df_dev)
data_words_test, essay_new_test = get_lemmetized(df_dev)

```

## Get Word_Embeddings

```python
def get_embedding_dict(path, vocab):
    """ Function to retreive embeddings from a file into a dictionary.
    @param path (str): path to the file where embeddings are to be written to.
    @param vocab (list): vocabulary list.
    @return embedding_dict (dict): Dictionary of words/documents as keys and their respective embeddings as values.
    """
    if path.split(".")[-1] == "bin":
        model = KeyedVectors.load_word2vec_format(path, binary=True, limit=300000)
    else:
        model = KeyedVectors.load_word2vec_format(path, binary=False)

    embedding_dict = dict()
    word_vectors = model.wv
    for word in vocab:
      try:
        if word in word_vectors.vocab :
          embedding_dict[word] = word_vectors[word.lower()]
          #print("im right",type(embedding_dict[word]))

       
      except: 
        print(word, 'not found')
        embedding_dict[word] = np.zeros((300,))
        #print("i'm wrong",embedding_dict[word])
        #print("i'm wrong",type(embedding_dict[word]))
      
        
    return embedding_dict
```

```python
def generate_embeddings(corpus, embedding_dict):
    """ Function to generate essay embeddings
    @param corpus (list): list of total words.
    @param embedding_dict (dict): word embeddings dictionary.
    """
    essay_embedding = []
    for word in corpus:
      temp = embedding_dict.get(word)
      if temp is None:
        temp = np.zeros((300,))
      essay_embedding.append(temp)
    return np.array(essay_embedding, dtype="float32")
    #return np.array([embedding_dict.get(text) for text in corpus.key])
```

## Get Synonyms

```python
def get_embedding_matrix1( path, vocab, top=1000000):
        """ Function to get the word embedding matrix.
        @param path (str): path to the word embeddings file.
        @param vocab (list): list of corpus vocab.
        @return embedding_matrix (np.array): embedding matrix.
        """
        if path.split(".")[-1] == "bin":
            embedding_model = KeyedVectors.load_word2vec_format(path, binary=True, limit=top)
            
        else:
            embedding_model = KeyedVectors.load_word2vec_format(path, binary=False, limit=top)
        #print(embedding_model)
        embeddin_model_vocab =[word for word in embedding_model.vocab.keys()]
        #print(embeddin_model_vocab)
        final_vocab = list(set(embeddin_model_vocab) | set(vocab))
        return final_vocab
```
